```python
n=int(input('enter number : '))
cnt=0
if n<2:
    print('Not prime')
elif n==2:
    print('Prime')
else:
    for i in range(2,n-1):
        if n%i==0:
            print(' not prime')
            cnt+=1
            break;
    if cnt==0:
        print('prime')
```

```python
n=input('enter a string : ')
i=0
j=len(n)-1
k=0
while i<j:
    if n[i]!=n[j]:
        print('Not palindrome')
        k+=1
        break
    else:
        i+=1
        j-=1
 
if k==0:
    print('palindrome')
```
