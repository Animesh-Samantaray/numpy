# Operation with advanced business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr=np.array([[1, 120000, 135000, 150000,160000],
    [2, 95000, 110000, 125000,230000],
    [3, 200000, 220000, 240000,560000],
    [4, 150000, 165000, 180000,123320],
    [5, 80000, 95000, 105000,879960]])
print('=== My sales ===')
# print(arr.shape)

# trying to get 1st 3 sample datas
print('sample data for 1st 3 resturants :\n',arr[:3,:5])#slicing
# print('sample data for 1st 3 resturants :\n',arr[:3])#slicing
```

```python
# To get the sum of sales of year

sum_list = np.sum(arr,axis=0)
print(sum_list)
# print the total sales of sum except id no
print(np.sum(arr[:,1:] ))
```

```python
#print the minimum sales of resturant
min_sales = np.min(arr[:,1:] , axis = 1)
print(f'minimum sales per resturant : {min_sales}')
```

```python
# maximum sales per year
max_sales = np.max(arr[:,1:],axis=0)
print(max_sales)
```

```python
#avg sales per resturant
avg_res = np.mean(arr[:,1:] , axis = 1)
print(avg_res)
```

### cumulative

```python
cumsum = np.cumsum(arr[:,1:],axis=1)
print(cumsum)

print(np.mean(cumsum , axis=0))
   
```

## Using MATPLOTLIB

```python
plt.figure(figsize=(10,6))
years = [2021, 2022, 2023, 2024]
plt.plot(years, np.mean(cumsum, axis=0))
plt.title('Average Cumulative Sales Across All Restaurants (20212025)')
plt.xlabel('Years')
plt.ylabel('Sales (in )')
plt.xticks(years)
plt.grid(True)
plt.tight_layout()
plt.show()

```
